```python
from pathlib import Path

from medminer.workflows.base.schema import DoctorsLetterState

docs = [
    DoctorsLetterState(patient_id=file.stem, letter=file.read_text())
    for file in (Path.cwd().parent.parent / "data" / "pre_medications").glob("*.txt")
]
```

```python
from medminer.utils.models import load_model
from medminer.workflows.medications.workflow import MedicationExtractionWorkflow

model = load_model()
workflow = MedicationExtractionWorkflow(model=model)
state = workflow.run_many(docs)
```
